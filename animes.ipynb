```python
df_trie = df.sort_values('Note_Globale', ascending=False)

print("Top 10 des meilleurs animés par note :")
print(df_trie[['Anime', 'Note_Globale', 'Note_Meilleur_Ep', 'Note_Pire_Ep', 'Nb_Episodes', 'Genre_Tags', 'Studio']].head(10).to_string(index=False))
```

```python
df_trie = df.sort_values('Note_Globale')

print("Top 10 des pires animés par note :")
print(df_trie[['Anime', 'Note_Globale', 'Note_Meilleur_Ep', 'Note_Pire_Ep', 'Nb_Episodes', 'Genre_Tags', 'Studio']].head(10).to_string(index=False))
```

```python
note_max = df['Note_Globale'].max()
note_min = df['Note_Globale'].min()

print(f"Note la plus haute : {note_max}")
print(f"Note la plus basse : {note_min}")


meilleur = df[df['Note_Globale'] == note_max]['Anime'].values[0]
print(f"Le mieux noté : {meilleur}")
```

```python
plt.figure(figsize=(10, 5))

plt.hist(df['Note_Globale'], bins=10, color='Pink', edgecolor='white')

moyenne = df['Note_Globale'].mean()
plt.axvline(moyenne, color='Purple', linestyle='--', linewidth=2, label=f'Moyenne: {moyenne:.1f}')

plt.title("Les Notes des Animés", fontsize=14, fontweight='bold')
plt.xlabel("Note")
plt.ylabel("Nombre d'animés")
plt.legend()
plt.show()

print("La ligne violette représente la moyenne des notes")
```

```python
plt.figure(figsize=(10, 6))

df_filtre = df[df['Nb_Episodes'] <= 500]

plt.scatter(df_filtre['Nb_Episodes'], df_filtre['Note_Globale'], 
           alpha=0.6, c='Pink', s=80, edgecolor='Black')

plt.title("Note vs Nombre d'épisodes", fontsize=14, fontweight='bold')
plt.xlabel("Nombre d'épisodes")
plt.ylabel("Note globale")
plt.grid(True, alpha=0.3)
plt.show()
```
