```python
import pandas as pd      
import matplotlib.pyplot as plt 
rose = "#f9a8d4"
lilas = "#c4b5fd"
bleu_pastel = "#93c5fd"
peche = "#fdba74"
menthe = "#a7f3d0"

plt.rcParams['figure.figsize'] = [10, 6]
plt.rcParams['axes.grid'] = True

print("‚úÖ Pandas et Matplotlib charg√©s !")
print("üéå Pr√™t √† analyser des anim√©s !")
```

```python
print("\nüìÇ √âTAPE 1 ‚Äî Chargement du fichier")
df = pd.read_csv("animes.csv")
print("‚úÖ Fichier charg√© avec succ√®s !")
print("Le tableau contient", df.shape[0], "lignes et", df.shape[1], "colonnes")
print(df.head(), "\n")
```

```python
print("üßº √âTAPE 2 ‚Äî Nettoyage des noms de colonnes")
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
print("Noms de colonnes propres ‚úÖ")
```

```python
print("üîÅ √âTAPE 3 ‚Äî Suppression des doublons")
df = df.drop_duplicates(subset="anime")
print("Doublons supprim√©s ‚úÖ")
print("Nouvelles dimensions :", df.shape, "\n")
```

```python
print("‚úèÔ∏è √âTAPE 4 ‚Äî Nettoyage du texte")
text_cols = [
    "anime", "genre_tags", "source", "status", "studio",
    "meilleur_ep_titre", "comm_meilleur_ep",
    "pire_ep_titre", "comm_pire_ep",
    "comm_saison_1", "comm_saison_2", "comm_saison_3"
]

for col in text_cols:
    df[col] = df[col].astype(str).str.strip()

df["anime"] = df["anime"].str.upper()

print("Texte nettoy√© et titres en majuscules ‚úÖ\n")

```

```python
print("üìÖ √âTAPE 5 ‚Äî Conversion des dates")
df["date_pub"] = pd.to_datetime(df["date_pub"], errors="coerce")
print("Dates converties ‚úÖ\n")
```

```python
print("‚≠ê √âTAPE 6 ‚Äî Notes limit√©es entre 0 et 10")
note_cols = ["note_globale", "note_meilleur_ep", "note_pire_ep"]
for col in note_cols:
    df[col] = df[col].clip(0, 10)

print("Notes v√©rifi√©es ‚úÖ\n")
```

```python
print("üé≠ √âTAPE 7 ‚Äî Transformation des genres")
df["genre_tags"] = df["genre_tags"].str.lower().str.replace(" ", "").str.split("/")
print("Genres transform√©s en liste ‚úÖ\n")

df["genre_principal"] = df["genre_tags"].apply(lambda x: x[0] if isinstance(x, list) and len(x) > 0 else None)

print("Genre principal extrait ‚úÖ\n")
```

```python
print("üì∫ √âTAPE 8 ‚Äî Gestion Saison 3")
df["comm_saison_3"] = df["comm_saison_3"].fillna("Pas de saison 3")
```

```python
print("üè∑Ô∏è √âTAPE 9 ‚Äî Colonnes en cat√©gories")
cat_cols = ["source", "status", "studio"]
for col in cat_cols:
    df[col] = df[col].astype("category")
print("Colonnes optimis√©es ‚úÖ\n")
```

```python
print("üîé √âTAPE 10 ‚Äî V√©rification dataset")

print("\n‚û°Ô∏è Nombre total de lignes :", len(df))

print("\n‚û°Ô∏è Colonnes qui ont encore des valeurs manquantes :")
print(df.isnull().sum())
print("Dataset clean üéâ\n")
```

```python
print("üèÜ √âTAPE 11 ‚Äî Analyse des notes et filtres \n")

df_trie = df.sort_values('note_globale', ascending=False)
print("üèÜ Top 10 des anim√©s par note :")
print(df_trie[['anime', 'note_globale', 'studio']].head(10).to_string(index=False))
```

```python
print("üìä √âTAPE 12 ‚Äî Statistiques g√©n√©rales")
note_moyenne = df['note_globale'].mean()
episodes_moyen = df['nb_episodes'].mean()
note_max = df['note_globale'].max()
note_min = df['note_globale'].min()

print(f"Note moyenne : {note_moyenne:.2f}")
print(f"√âpisodes moyens : {episodes_moyen:.0f}")
print(f"Note max : {note_max}")
print(f"Note min : {note_min}")

meilleur = df[df['note_globale'] == note_max]['anime'].values[0]
print(f"üèÜ Meilleur anim√© : {meilleur}\n")
```

```python
condition1 = df['note_globale'] > 8
condition2 = df['status'] == 'Fini'
bons_finis = df[condition1 & condition2]
print(f"\n‚úÖ {len(bons_finis)} anim√©s avec note > 8 ET termin√©s :")
print(bons_finis[['anime', 'note_globale', 'status']].head(10).to_string(index=False))
```

```python
print(" √âTAPE 12 ‚Äî Statistiques et visualisations")

note_moyenne = df['note_globale'].mean()
print(f"\nüìä Note moyenne des anim√©s : {note_moyenne:.2f} / 10")
```

```python
episodes_moyen = df['nb_episodes'].mean()
print(f"üì∫ Nombre moyen d'√©pisodes : {episodes_moyen:.0f}")
```

```python
note_max = df['note_globale'].max()
note_min = df['note_globale'].min()
print(f"‚≠ê Note la plus haute : {note_max}")
print(f"üìâ Note la plus basse : {note_min}")
```

```python
statuts = df['status'].value_counts()
print("\nüìä R√©partition par statut :")
print(statuts)
```

```python
print("üè¢ Top 5 des studios : \n")
print(df['studio'].value_counts().head(5)) # Affiche les 5 studios produisant le plus d‚Äôanim√©s
```

```python
print("\nüìä √âTAPE 13 ‚Äî Graphique distribution des notes \n")
plt.figure(figsize=(8,5))
plt.hist(df['note_globale'], bins=10, color=lilas, edgecolor="white")
plt.axvline(note_moyenne, color=rose, linestyle='--', linewidth=3, label="Moyenne")
plt.title("Distribution des notes")
plt.show()
```

```python
statuts = df['status'].value_counts()
plt.figure(figsize=(8,5)) 
bars = plt.bar(statuts.index, statuts.values, color=[rose, lilas, bleu_pastel, peche, menthe])

plt.title("Anim√©s par statut")
plt.show()
```

```python
df_filtre = df[df['nb_episodes'] <= 500]
plt.figure(figsize=(8,5))
plt.scatter(df_filtre['nb_episodes'], df_filtre['note_globale'],
            alpha=0.7, c=lilas, s=90, edgecolor= "purple")
plt.title("Note vs Nombre d'√©pisodes")
plt.xlabel("√âpisodes")
plt.ylabel("Note")
plt.show()
```
