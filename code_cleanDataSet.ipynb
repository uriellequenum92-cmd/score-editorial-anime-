```python
print("\nüìÇ √âTAPE 1 ‚Äî Chargement du fichier")
df = pd.read_csv("animes.csv")
print("‚úÖ Fichier charg√© avec succ√®s !")
print("Le tableau contient", df.shape[0], "lignes et", df.shape[1], "colonnes")
print(df.head(), "\n")
```

```python
print("üßº √âTAPE 2 ‚Äî Nettoyage des noms de colonnes")
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
print("Noms de colonnes propres ‚úÖ")
```

```python
print("üîÅ √âTAPE 3 ‚Äî Suppression des doublons")
df = df.drop_duplicates(subset="anime")
print("Doublons supprim√©s ‚úÖ")
print("Nouvelles dimensions :", df.shape, "\n")
```

```python
print("‚úèÔ∏è √âTAPE 4 ‚Äî Nettoyage du texte")
text_cols = [
    "anime", "genre_tags", "source", "status", "studio",
    "meilleur_ep_titre", "comm_meilleur_ep",
    "pire_ep_titre", "comm_pire_ep",
    "comm_saison_1", "comm_saison_2", "comm_saison_3"
]

for col in text_cols:
    df[col] = df[col].astype(str).str.strip()

df["anime"] = df["anime"].str.upper()

print("Texte nettoy√© et titres en majuscules ‚úÖ\n")

```

```python
print("üìÖ √âTAPE 5 ‚Äî Conversion des dates")
df["date_pub"] = pd.to_datetime(df["date_pub"], errors="coerce")
print("Dates converties ‚úÖ\n")
```

```python
print("‚≠ê √âTAPE 6 ‚Äî Notes limit√©es entre 0 et 10")
note_cols = ["note_globale", "note_meilleur_ep", "note_pire_ep"]
for col in note_cols:
    df[col] = df[col].clip(0, 10)

print("Notes v√©rifi√©es ‚úÖ\n")
```

```python
print("üé≠ √âTAPE 7 ‚Äî Transformation des genres")
df["genre_tags"] = df["genre_tags"].str.lower().str.replace(" ", "").str.split("/")
print("Genres transform√©s en liste ‚úÖ\n")

df["genre_principal"] = df["genre_tags"].apply(lambda x: x[0] if isinstance(x, list) and len(x) > 0 else None)

print("Genre principal extrait ‚úÖ\n")
```

```python
print("üì∫ √âTAPE 8 ‚Äî Gestion Saison 3")
df["comm_saison_3"] = df["comm_saison_3"].fillna("Pas de saison 3")
```

```python
print("üè∑Ô∏è √âTAPE 9 ‚Äî Colonnes en cat√©gories")
cat_cols = ["source", "status", "studio"]
for col in cat_cols:
    df[col] = df[col].astype("category")
print("Colonnes optimis√©es ‚úÖ\n")
```

```python
print("\n‚û°Ô∏è Aper√ßu des 5 premi√®res lignes :")
print(df.head())

print("\n‚û°Ô∏è Nombre total de lignes :", len(df))

print("\n‚û°Ô∏è Colonnes qui ont encore des valeurs manquantes :")
print(df.isnull().sum())
```

```python
print("üíæ √âTAPE 11 ‚Äî Sauvegarde du fichier")
df.to_csv("animes_clean.csv", index=False)
print("‚úÖ Fichier sauvegard√© : animes_clean.csv")
```
